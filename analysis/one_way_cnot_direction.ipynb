```python
from qiskit import QuantumCircuit
from qiskit_ibm_provider import IBMProvider

import pickle
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from models.victim_params import VictimParams
from qiskit.result import marginal_counts
```

```python
provider = IBMProvider()
```

# IBM Nairobi

## Parameters

```python
job_id_3_5 = 'ch4p9h05ngm0qhemvrb0'
job_id_5_3 = 'ch4pd3o1n9kgjggblvp0'

snooper_qubits = (1, 4, 6)
victim_qubits = (3, 5)
ancilla_qubits = (0, 2)
victim_circ_uuid = '3a03572b-c786-449e-b9ed-4fea2fb16b54'
```

## Fetch data from backend

```python
job_3_5 = provider.backend.retrieve_job(job_id_3_5)
job_5_3 = provider.backend.retrieve_job(job_id_5_3)
```

```python
job_3_5.result().backend_name == job_5_3.result().backend_name
```

## Analysis

```python
for creg_index, snooper_qubit in enumerate(snooper_qubits):
    counts_3_5 = pd.DataFrame(marginal_counts(job_3_5.result(), [creg_index]).get_counts())['0']
    counts_5_3 = pd.DataFrame(marginal_counts(job_5_3.result(), [creg_index]).get_counts())['0']
    fig, ax = plt.subplots(layout='constrained', figsize=(12, 6))
    ax.set_title('0 counts for qubit {}'.format(snooper_qubit))
    ax.plot(counts_3_5, label='counts_3_5')
    ax.plot(counts_5_3, label='counts_5_3', alpha=0.7)
    period = 60
    for i in range(period, len(counts_3_5)+1, period):
        ax.axvline(i, color='red', linestyle='--')
    ax.legend()
    plt.show()
```

```python
for creg_index, snooper_qubit in enumerate(snooper_qubits):
    print(snooper_qubit)
    counts_3_5 = pd.DataFrame(marginal_counts(job_3_5.result(), [creg_index, ]).get_counts())['0']
    counts_5_3 = pd.DataFrame(marginal_counts(job_5_3.result(), [creg_index, ]).get_counts())['0']
    diff = counts_3_5 - counts_5_3
    x_range = np.arange(len(counts_3_5))
    fig, ax = plt.subplots(layout='constrained', figsize=(12, 6))
    ax.set_title('0 counts difference for qubit {}'.format(snooper_qubit))
    ax.scatter(x_range, diff, label='counts_3_5-counts_5_3')
    period = 60
    for i in range(period, len(counts_3_5)+1, period):
        ax.axvline(i, color='red', linestyle='--')
    ax.axhline(y=0, color='k')
    plt.show()
```

```python
for creg_index, snooper_qubit in enumerate(snooper_qubits):
    counts_3_5 = pd.DataFrame(marginal_counts(job_3_5.result(), [creg_index, ]).get_counts())['0']
    counts_5_3 = pd.DataFrame(marginal_counts(job_5_3.result(), [creg_index, ]).get_counts())['0']
    diff = counts_3_5 - counts_5_3
    diff2 = diff.iloc[0:240]
    x_range = np.arange(0, 240)
    print(len(diff2))
    fig, ax = plt.subplots(layout='constrained', figsize=(12, 6))
    ax.set_title('0 counts difference for qubit {}'.format(snooper_qubit))
    ax.scatter(x_range, diff2, label='counts_3_5-counts_5_3')
    period = 60
    for i in range(period, len(diff2)+1, period):
        ax.axvline(i, color='red', linestyle='--')
    ax.axhline(y=0, color='k')
    plt.show()
```

```python
for creg_index, snooper_qubit in enumerate(snooper_qubits):
    counts_3_5 = pd.DataFrame(marginal_counts(job_3_5.result(), [creg_index, ]).get_counts())['0']
    counts_5_3 = pd.DataFrame(marginal_counts(job_5_3.result(), [creg_index, ]).get_counts())['0']
    diff = counts_3_5 - counts_5_3
    diff2 = diff.iloc[240:480]
    x_range = np.arange(240, 480)
    print(len(diff2))
    fig, ax = plt.subplots(layout='constrained', figsize=(12, 6))
    ax.set_title('0 counts difference for qubit {}'.format(snooper_qubit))
    ax.scatter(x_range, diff2, label='counts_3_5-counts_5_3')
    period = 60
    for i in range(240+period, 240+len(diff2)+1, period):
        ax.axvline(i, color='red', linestyle='--')
    ax.axhline(y=0, color='k')
    plt.show()
```
